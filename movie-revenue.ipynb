```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
data=pd.read_csv('movies.csv')
```

```python
data.head()
```

```python
x=data['Production Budget ($)'].values
y=data['Worldwide Gross ($)'].values
```

```python
np.corrcoef(x,y)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(x,y,color='purple',label='points')
plt.title('movie revolution')
plt.xlabel('Production Budget ($)')
plt.ylabel('Worldwide Gross ($)')
plt.xlim(0,450000000)
plt.ylim(0,3000000000)
plt.legend()
plt.show()
```

```python
meanOfX=np.mean(x)
meanOfY=np.mean(y)
```

```python
length=len(x)
```

```python
numerator=0
denominator=0
```

```python
for i in range (length):
    numerator+=(x[i]-meanOfX)*(y[i]-meanOfY)
    denominator+=(x[i]-meanOfX)**2
```

```python
m=numerator/denominator
```

```python
c=meanOfY-(m*meanOfX)
```

```python
Y=x*m+c
plt.figure(figsize=(10,7))
plt.scatter(x,y,color='purple',label='points')
plt.plot(x,Y,color='green',label='perfect line',linewidth=4)
plt.xlim(0,450000000)
plt.ylim(0,3000000000)
plt.title('movie revolution')
plt.xlabel('Production Budget ($)')
plt.ylabel('Worldwide Gross ($)')
plt.legend()

plt.show()
```

```python

Y_pred=0
Y_act=0
for i in range(length):
    y_p=m*x[i]+c
   
    Y_pred+=(y[i]-y_p)**2
    Y_act+=(y[i]-meanOfY)**2
r2=1-(Y_pred/Y_act)
```
